```python
from social_media import Facebook

# Add your Facebook token here
token = ''
fb = Facebook(token)
```

```python
fb.get_page_info('jdoriajr')
```

```python
recent_posts = fb.page_post_list('jdoriajr', days_past=7)
recent_posts
```

```python
fb.post_stats(recent_posts[2])
```

```python
import pandas as pd

posts = []
for post in recent_posts:
    details = fb.post_stats(post)
    posts.append(details)

df = pd.DataFrame(posts)
df.head()
```

```python
df.to_csv('jdoria_facebook.csv')
```
